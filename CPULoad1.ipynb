```python
def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        print("Train Result:\n")
        print("score : {}".format(clf.score(X_train, y_train)))
        print("R2_score : {}".format(r2_score(y_train,clf.predict(X_train))))
        print("mean_squared_error : {}".format(m.sqrt(mean_squared_error(clf.predict(X_train), y_train))))
        return ()
        
    elif train==False:
        print("Test Result:\n") 
        print("score : {}".format(clf.score(X_test, y_test)))
        print("R2_score : {}".format(r2_score(y_test,clf.predict(X_test))))
        print("mean_squared_error : {}".format(m.sqrt(mean_squared_error(clf.predict(X_test), y_test))))
       
```

```python
train_sizes, train_scores, test_scores = learning_curve(load_xgb_model6, 
                                                        Xt, 
                                                        yt,
                                                        cv=5,
                                                        scoring='r2',
                                                        n_jobs=-1, 
                                                        verbose=10,
                                                        train_sizes=np.linspace(0.01, 1.0, 10))
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
```

```python
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

```python
plt.plot(train_sizes, train_mean, '--', color="#111111",  label="Training score")
plt.plot(train_sizes, test_mean, color="#111111", label="Cross-validation score")
```
